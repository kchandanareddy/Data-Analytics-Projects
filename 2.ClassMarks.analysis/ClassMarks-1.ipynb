# Description

The dataset consists of student marks across various questions in an exam. It contains 86 records with 12 columns, where each row represents a student's total score and their marks for different questions. The columns include:

Total: The overall score obtained by each student.
Q1aM4 to Q6bM6: Marks obtained in specific question parts, with varying maximum scores (e.g., "Q1aM4" indicates a question with a maximum of 4 marks).
The dataset exhibits missing values in several columns, indicating that not all students attempted every question. The missing values are particularly significant in later questions, suggesting either a drop in student responses or incomplete evaluations.

```python
import pandas as pd
a=pd.read_csv("class_marks.csv")

def assign_grade(Total):
    if Total >= 35:
        return 'A'
    elif Total >= 30:
        return 'B'
    elif Total >= 20:
        return 'C'
    elif Total >= 15:
        return 'D'
    else:
        return 'F'

a['Grade'] = a['Total'].apply(assign_grade)

a
```

```python
a=a.loc[(a['Total'] >= 15) & (a['Total'] <= 20)]
a=a.reset_index()
a
```
